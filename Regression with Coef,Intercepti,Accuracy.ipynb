```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df=pd.read_csv("student.csv")
```

```python
df.columns=["sinif","saat","puan"]
```

```python
df.head(3)
```

```python
# @title saat vs puan

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='saat', y='puan', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
y=df[["puan"]]
x=df[["sinif","saat"]]
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

```python
model.coef_ #katsayi
```

```python
model.intercept_ #sabit
```

```python
model.predict([[4,5]])
```

```python
model.predict([[3,4.508]])
```
